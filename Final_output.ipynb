```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
#imporing csv
boston_train = pd.read_csv("/content/train.csv")
```

```python
#reading the features
boston_train.head()
```

```python
#datatypes and length of the features
boston_train.info()
```

```python
#checking NULL values
boston_train.isna().sum()
```

```python
#splitting into features into train anf target
X = boston_train.iloc[:,1:14].values
Y = boston_train.iloc[:,-1].values
LR = LinearRegression()
LR.fit(X,Y)
print("Intercept : ", LR.intercept_)
print("Coefficients : ", LR.coef_)
```

```python
#Checking the minimum values
print(boston_train.min())
```

```python
#Checking the errors
Y_pred = LR.predict(X)
RMSE = np.sqrt(mean_squared_error(Y,Y_pred))
print("Root Mean Square Error : ", RMSE)
```

```python
#Checking for outliers
fig, axs = plt.subplots(nrows=5,ncols=3,figsize=(10,8))
index = 0
axs = axs.flatten()
for k,v in boston_train.items():
    sns.boxplot(y=k,data=boston_train,ax=axs[index])
    index+=1
plt.tight_layout()
```

```python
# Removing Outliers
boston_train = boston_train[~((boston_train['crim']>=25))]
boston_train = boston_train[~((boston_train['rm']>=8.5)|(boston_train['rm']<4))]
boston_train = boston_train[~((boston_train['lstat']>=30))]
```

```python
#Training again
X = boston_train.iloc[:,1:14].values
Y = boston_train.iloc[:,-1].values
LR = LinearRegression()
LR.fit(X,Y)
print("Intercept : ", LR.intercept_)
print("Coefficients : ", LR.coef_)
```

```python
#Checking again
Y_pred = LR.predict(X)
RMSE = np.sqrt(mean_squared_error(Y,Y_pred))
print("Root Mean Square Error : ", RMSE)
```

```python
# Correlation Matric for Feature Selection
plt.figure(figsize = (20,10))
cmap = sns.diverging_palette(500,10,as_cmap=True)
sns.heatmap(boston_train.corr(),cmap=cmap,center=0,square=True,annot=True)
```

```python
#Feature selction
X = boston_train[['lstat','rm','indus','ptratio','tax','nox','crim','age']].values
Y = boston_train[['medv']].values
```

```python
#Training again
LR = LinearRegression()
LR.fit(X,Y)
print("Intercept : ", LR.intercept_)
print("Coefficients : ", LR.coef_)
#Checking again
Y_pred = LR.predict(X)
RMSE = np.sqrt(mean_squared_error(Y,Y_pred))
print("Root Mean Square Error : ", RMSE)
```

```python
# Prediction for Test Dataset
boston_test = pd.read_csv('/content/test.csv')
X_test = boston_test[['lstat','rm','indus','ptratio','tax','nox','crim','age']].values
Y_test_pred = LR.predict(X_test)
for x, y in zip(boston_test['ID'],Y_test_pred):
    print(x, y, sep='\t\t')
```
